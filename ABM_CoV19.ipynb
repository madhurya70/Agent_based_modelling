```python
import osmnx as ox
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gdf = ox.geometries.geometries_from_bbox(52.6500, 52.5900,-1.0500,-1.1600, tags = {'building':True})
gdf1 = ox.geometries.geometries_from_bbox(52.6500, 52.5900,-1.0500,-1.1600, tags = {'amenity':True})
```

```python
lat_household=[]
long_household =[]
new_abc1 = pd.read_csv('abc1.csv')
for i in range(5194):
   if(new_abc1['building'][i]=='yes' or new_abc1['building'][i]=='apartments' or new_abc1['building'][i]=='residential'):
        x = new_abc1['geometry'][i]
        word = ""
        for e in x:
            word+= str(e)
        list1 =[]   
        list1 = word.split(",")  
        long_household.append(float(list1[0]))
        lat_household.append(float(list1[1]))
lat_retail=[]
long_retail =[]
new_abc1 = pd.read_csv('abc1.csv')
for i in range(5194):
   if(new_abc1['building'][i]=='retail' or new_abc1['building'][i]=='commercial'):
        x = new_abc1['geometry'][i]
        word = ""
        for e in x:
            word+= str(e)
        list1 =[]   
        list1 = word.split(",")  
        long_retail.append(float(list1[0]))
        lat_retail.append(float(list1[1]))
lat_offices=[]
long_offices =[]
new_abc1 = pd.read_csv('abc1.csv')
for i in range(5194):
   if(new_abc1['building'][i]=='office' or new_abc1['building'][i]=='industrial' or new_abc1['building'][i]=='mill' ):
        x = new_abc1['geometry'][i]
        word = ""
        for e in x:
            word+= str(e)
        list1 =[]   
        list1 = word.split(",")  
        long_offices.append(float(list1[0]))
        lat_offices.append(float(list1[1]))
lat_University=[]
long_University =[]
new_abc1 = pd.read_csv('abc1.csv')
for i in range(5194):
   if(new_abc1['building'][i]=='university'):
        x = new_abc1['geometry'][i]
        word = ""
        for e in x:
            word+= str(e)
        list1 =[]   
        list1 = word.split(",")  
        long_University.append(float(list1[0]))
        lat_University.append(float(list1[1]))
len(lat_University)
```

```python
lat_restaurants=[]
long_restaurants =[]
new_abc2 = pd.read_csv('abc2.csv')
for i in range(len(new_abc2['amenity'])):
   if(new_abc2['amenity'][i]=='restaurant' or new_abc2['amenity'][i]=='fast_food' or new_abc2['amenity'][i]=='pub'):
        x = new_abc1['geometry'][i]
        word = ""
        for e in x:
            word+= str(e)
        list1 =[]   
        list1 = word.split(",")  
        long_restaurants.append(float(list1[0]))
        lat_restaurants.append(float(list1[1]))

lat_hospitals =[]
long_hospitals =[]
for i in range(len(new_abc2['amenity'])):
   if(new_abc2['amenity'][i]=='hospital'):
        x = new_abc1['geometry'][i]
        word = ""
        for e in x:
            word+= str(e)
        list1 =[]   
        list1 = word.split(",")  
        long_hospitals.append(float(list1[0]))
        lat_hospitals.append(float(list1[1]))
lat_schools =[]
long_schools =[]
for i in range(len(new_abc2['amenity'])):
   if(new_abc2['amenity'][i]=='school' or new_abc2['amenity'][i]=='college'):
        x = new_abc1['geometry'][i]
        word = ""
        for e in x:
            word+= str(e)
        list1 =[]   
        list1 = word.split(",")  
        long_schools.append(float(list1[0]))
        lat_schools.append(float(list1[1]))
lat_worship =[]
long_worship =[]
for i in range(len(new_abc2['amenity'])):
   if(new_abc2['amenity'][i]=='church' or new_abc2['amenity'][i]=='place_of_worship'):
        x = new_abc1['geometry'][i]
        word = ""
        for e in x:
            word+= str(e)
        list1 =[]   
        list1 = word.split(",")  
        long_worship.append(float(list1[0]))
        lat_worship.append(float(list1[1]))
len(lat_worship)
```

```python
x_lat = np.zeros(shape =(5194))
x_long = np.zeros(shape =(5194))
for i in range(len(long_household)):
    x_long[i] = (float(long_household[i]) +1)
for i in range(len(long_household)):
    x_lat[i] = (float(lat_household[i]) -52)
x_lat1 = np.zeros(shape =(403))
x_long1 = np.zeros(shape =(403))
for i in range(len(long_restaurants)):
    x_long1[i] = (float(long_restaurants[i]) +1)
for i in range(len(long_restaurants)):
    x_lat1[i] = (float(lat_restaurants[i]) -52)
x_lat4 = np.zeros(shape =(8))
x_long4 = np.zeros(shape =(8))
for i in range(len(long_hospitals)):
    x_long4[i] = (float(long_hospitals[i]) +1)
for i in range(len(long_hospitals)):
    x_lat4[i] = (float(lat_hospitals[i]) -52)
x_lat5 = np.zeros(shape =(208))
x_long5 = np.zeros(shape =(208))
for i in range(len(long_retail)):
    x_long5[i] = (float(long_retail[i]) +1)
for i in range(len(long_retail)):
    x_lat5[i] = (float(lat_retail[i]) -52)
x_lat6 = np.zeros(shape =(307))
x_long6 = np.zeros(shape =(307))
for i in range(len(long_offices)):
    x_long6[i] = (float(long_offices[i]) +1)
for i in range(len(long_offices)):
    x_lat6[i] = (float(lat_offices[i]) -52)
x_lat7 = np.zeros(shape =(307))
x_long7 = np.zeros(shape =(307))
for i in range(len(long_University)):
    x_long7[i] = (float(long_University[i]) +1)
for i in range(len(long_University)):
    x_lat7[i] = (float(lat_University[i]) -52)
x_lat8 = np.zeros(shape =(93))
x_long8 = np.zeros(shape =(93))
for i in range(len(long_schools)):
    x_long8[i] = (float(long_schools[i]) +1)
for i in range(len(long_schools)):
    x_lat8[i] = (float(lat_schools[i]) -52)
x_lat9 = np.zeros(shape =(181))
x_long9 = np.zeros(shape =(181))
for i in range(len(long_worship)):
    x_long9[i] = (float(long_worship[i]) +1)
for i in range(len(long_worship)):
    x_lat9[i] = (float(lat_worship[i]) -52)
```

```python
from matplotlib.axis import Axis 
plt.figure(figsize=(20,20))
plt.plot(x_lat, x_long, '*', label = "buildings/apartments")
plt.plot(x_lat1, x_long1, '*', label ="restaurants/fast_food")
plt.plot(x_lat4, x_long4, '*', label ="hospitals")
plt.plot(x_lat5, x_long5, '*', label ="retail/commercial")
plt.plot(x_lat6, x_long6, '*', label ="office/industrial/mill")
plt.plot(x_lat7, x_long7, '*', label ="Univesity")
plt.plot(x_lat8, x_long8, '*', label ="Schools/colleges")
plt.plot(x_lat9, x_long9, '*', label ="places of worship")
plt.legend()
plt.xlim(0.58, 0.66)
plt.ylim(-0.04, -0.17)
```
